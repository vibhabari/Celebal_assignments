```python
#HANDLE MISSING VALUES

print(df.isnull().sum())
```

```python
df['discount']=df["discount"].fillna(df["discount"].mean())
df['what_customers_said']    = df['what_customers_said'].fillna("No review")
df['seller_name'] = df['seller_name'].fillna(df['seller_name'].mode()[0])
df['videos']   = df['videos'].fillna("No video")
df['seller_information']    = df['seller_information'].fillna("No available")
df['variations'] = df['variations'].fillna(df['variations'].mode()[0])

print('Missing values per column (after):')
print(df.isnull().sum())
```

```python
#FILTER ROWS AND SELECT COLUMNS
filtered_df = df[df["rating"] > 4][['product_id','title','rating','final_price']]
print("products with rating > 4\n")
print(filtered_df.head())
```

```python
discount_products = df[df["discount"] > 50][['product_id','title','initial_price','discount','final_price']]
print("Products with discount greater than 50\n")
print(discount_products.head())
```

```python
# CHECK FOR DUPLICATE VALUES
print(df.duplicated().sum())
```

no duplicates found


```python
#DERIVE NEW COLUMN : DISCOUNT AMOUNT
df["final_price"] = (
    df["final_price"]
    .astype(str)
    .str.replace('₹', '')
    .str.replace(',', '')
    .str.replace('"', '')
)

df["final_price"] = pd.to_numeric(df["final_price"])

df["discount_amount"] = (df["initial_price"] - df["final_price"])


print( df[ [ "title", "initial_price", "final_price", "discount_amount"]].head(50))
```
