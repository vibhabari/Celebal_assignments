```python
# START SPARK SESSION
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *


spark = SparkSession.builder.appName("Spark Assignment").getOrCreate()
```

```python
# Read CSV file
df_csv = (spark.read.option("header", True).option("inferSchema", True).option("quote", '"').option("escape", '"')
.option("multiLine", True).csv("/Volumes/workspace/default/datasets/Superstore.csv"))
df_csv.printSchema()
```

```python
# Read Parquet file
df_parquet = (spark.read.parquet("/Volumes/workspace/default/datasets/Superstore.parquet"))
df_parquet.printSchema()
```

```python
df_csv.show(10)
```

```python
df_parquet.show(5)
```

```python
# COLUMN SELECTION & FILTERING
# 1. Select Product ID and Sales where Category is Technology
df_csv.select("Product ID", "Sales") .filter(col("Category") == "Technology") .show(5)

# 2. Filter Category = Technology AND Sales > 1000
df_csv.filter((col("Category") == "Technology") & (col("Sales") > 1000)).show(5)

# 3. Filter Region = East OR Category = Technology
df_csv.filter( (col("Region") == "East") | (col("Category") == "Technology")).show(5)
```

```python
# COLUMN SELECTION & RENAMING ON CSV
print("\n COLUMN SELECTION & RENAMING")

df_selected = (df_csv.select("Order ID", "Order Date", "Ship Date", "Ship Mode","Customer Name", "Segment", "Region", "State", "City","Category", "Sub-Category", "Product Name","Sales", "Quantity", "Discount", "Profit")
.withColumnRenamed("Sub-Category", "Sub_Category")
.withColumnRenamed("Customer Name", "Customer_Name")
.withColumnRenamed("Product Name", "Product_Name")
.withColumnRenamed("Order ID", "Order_ID")
.withColumnRenamed("Order Date", "Order_Date")
.withColumnRenamed("Ship Date", "Ship_Date")
.withColumnRenamed("Ship Mode", "Ship_Mode"))

print("  Selected & renamed columns:")
df_selected.printSchema()
```

```python
# COLUMN SELECTION & RENAMING ON PARQUET
print("\n COLUMN SELECTION & RENAMING")

df_selected1 = (df_parquet.select("Order ID", "Order Date", "Ship Date", "Ship Mode","Customer Name", "Segment", "Region", "State", "City","Category", "Sub-Category", "Product Name","Sales", "Quantity", "Discount", "Profit")
.withColumnRenamed("Sub-Category", "Sub_Category")
.withColumnRenamed("Customer Name", "Customer_Name")
.withColumnRenamed("Product Name", "Product_Name")
.withColumnRenamed("Order ID", "Order_ID")
.withColumnRenamed("Order Date", "Order_Date")
.withColumnRenamed("Ship Date", "Ship_Date")
.withColumnRenamed("Ship Mode", "Ship_Mode"))

print("  Selected & renamed columns:")
df_selected1.printSchema()
```

```python
# COLUMN TRANSFORMATIONS
df_transformed = (df_selected
# Cast date strings → proper DateType
.withColumn("Order_Date",to_date("Order_Date", "M/d/yyyy"))
.withColumn("Ship_Date",to_date("Ship_Date",  "M/d/yyyy"))

# Derived: Final_sales
.withColumn("Final_Sales",round(col("Sales") * 1.18, 2))

# Derived: profit tier label
.withColumn("Profit_Tier",when(col("Profit") > 100,  "High")
.when(col("Profit") > 0,    "Positive")
.when(col("Profit") == 0,   "Break-Even")
.otherwise("Loss")))

print("  Transformed schema:")
df_transformed.printSchema()
print("  Sample rows with new columns:")
df_transformed.select("Order_Date","Ship_Date","Final_Sales","Profit_Tier").show(8, truncate=20)
```

```python
#WIDE TRANSFORMATIONS
print("\nWIDE TRANSFORMATIONS")
#  Profit Tier distribution
profittier_dist = (df_transformed.groupBy("Profit_Tier")
.agg(count("*").alias("Order_Count"), round(sum("Profit"), 2).alias("Total_Profit"), round(sum("Sales"),2).alias("Total_Sales"))
.orderBy(col("Order_Count").desc()))
print("  Profit Tier distribution:")
profittier_dist.show(truncate=False)

#  Top 10 sub-categories by revenue
top_subcat = (df_transformed.groupBy("Category", "Sub_Category")
.agg(round(sum("Sales"),2).alias("Revenue"), round(sum("Profit"),2).alias("Profit"), count("*").alias("Txns"))
.orderBy(col("Revenue").desc())
.limit(10))
print("  Top 10 Sub-Categories by Revenue:")
top_subcat.show(truncate=False)
```

```python
#  Predicate pushdown
filtered_df = df_parquet.filter(col("Sales") > 1000)
filtered_df.show()
```

```python
# NULL VALUE HANDLING
print("\n NULL VALUE HANDLING")

null_counts = (df_transformed.select([sum(col(c).isNull().cast("int")).alias(c)
for c in df_transformed.columns]))

print("  Null counts per column:")
null_counts.show(truncate=False)

# Drop rows where Sales or Profit is null 
df_clean = df_transformed.dropna(subset=["Sales", "Profit"])
print(f"  Rows before null drop : {df_transformed.count():,}")
print(f"  Rows after  null drop : {df_clean.count():,}")
```

```python
df_clean.write.mode("overwrite").option("header", True).csv("/Volumes/workspace/default/datasets/outputs_csv")

```

```python
# Complete data pipelines (read → transform → filter → write).

# Read
df_pipeline = (spark.read.option("header", True).option("inferSchema", True).option("quote", '"').option("escape", '"')
.option("multiLine", True).csv("/Volumes/workspace/default/datasets/Superstore.csv"))
df_pipeline.printSchema()

# Transform
df_pipeline = (df_pipeline.withColumnRenamed("Product Name", "Product_Name").withColumn("Final_Sales", round(col("Sales") * 1.18, 2)))

# Filter
df_pipeline = df_pipeline.filter((col("Sales") > 500) &(col("Profit") > 0))

# Write
df_pipeline.write.mode("overwrite").option("header", True).csv("/Volumes/workspace/default/datasets/pipeline_output_csv")

```
